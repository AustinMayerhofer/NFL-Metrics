```python
import pandas as pd  # This is always assumed but is included here as an introduction.
import numpy as np
import matplotlib.pyplot as plt

BACKSLASH_FOLLOWING_REGEX = '(?<=\\\\).+' # regex for anything after a backslash

df = pd.read_csv('../data/years/2021/2021_Player_Receiving.csv')

# Add column [PlayerID]
df['PlayerID'] = df['Player'].str.extract('(' + BACKSLASH_FOLLOWING_REGEX + ')')

# [Player] cleaning -- remove pro bowl/all-pro designations, remove backslash + player ID
df = df.replace(BACKSLASH_FOLLOWING_REGEX, '', regex=True)
df = df.replace('\\\\', '', regex=True)
df = df.replace('\*', '', regex=True) # PFR uses * for pro bowl
df = df.replace('\+', '', regex=True) # PFR uses + for all-pro

# [Rk] drop -- unnecessary
df = df.drop('Rk', 1)

# [Ctch%] cleaning -- convert from object to float type
df['Ctch%'] = df['Ctch%'].str.rstrip('%').astype('float') / 100.0

df
```
